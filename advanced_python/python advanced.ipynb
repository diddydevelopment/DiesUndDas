## Decorateure

```python
def debug_decor(func):
    def decorate(*args,**kwargs):
        print('debug call ',func,args,kwargs)
        return func(*args,**kwargs)
    return decorate

@debug_decor
def do_this(a,b):
    return b+a+1   
    
@debug_decor
def do_that(a,b):
    return a+b

print(do_this(5,6))
print(do_that(5,6))
```

## Decorateure mit Parameter

```python
def repeat_decorator(x):
    def decorator(func):
        def decorate(*args,**kwargs):
            for i in range(x):
                func(*args,**kwargs)
        return decorate
    return decorator

@repeat_decorator(5)
def print_that(a):
    print(a)

print_that('hallo welt')
    
```

```python
class MyRange:
    def __init__(self,a,b=0,step=1):
        if b:
            b,a = a,b
        self.lower = b
        self.upper = a
        self.step = step
        self.current = self.lower
    def __iter__(self):
        return self
    def __next__(self):
        old = self.current
        if self.current >= self.upper: raise StopIteration
        self.current += self.step
        return old

for x in MyRange(4,8):
    print(x)

mr = MyRange(4,8)
    
while(True):
    try:
        print(mr.__next__())
    except StopIteration:
        break
    
```

## Generatoren

```python
import random
def random_gen(mi,ma,num=10):
    for i in range(num):
        yield random.random()*(ma-mi)+mi

for r in random_gen(3.4,8.9):
    print(r)
```

```python
def range_gen(a,b,step=1):
    if b:
        b,a = a,b
    lower = b
    upper = a
    current = lower
    while(True):
        if current >= upper:
            break
        old = current
        current += step
        yield old

for r in range_gen(4,8,2):
    print(r)
```

```python
class withi:
    def __init__(self):
        print('init')
    def __enter__(self):
        print('enter')
        return self
    def __exit__(self,ex_type,ex_value,ex_trace):
        print('exit',ex_type,ex_value,ex_trace)
    def say_hello(self):
        print('hello')

with withi() as w:
    print('withiiii')
    w.say_hello()
```
